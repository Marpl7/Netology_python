# Выбросы

```python
q1 = horses_new.rectal_temperature.quantile(0.25)
q3 = horses_new.rectal_temperature.quantile(0.75)
print(q1, q3)
iqr = q3 - q1
lower_bound = q1 - (2 * iqr) 
upper_bound = q3 + (2 * iqr)
remove_outliers = horses_new[~horses_new.rectal_temperature.between(lower_bound, upper_bound, inclusive=True)].sort_values('rectal_temperature')
remove_outliers
```

```python
q1 = horses_new.Puls.quantile(0.25)
q3 = horses_new.Puls.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - iqr 
upper_bound = q3 + (1.5 * iqr)
print(lower_bound, upper_bound)
remove_outliers = horses_new[~horses_new.Puls.between(lower_bound, upper_bound, inclusive=True)].sort_values('Puls')
remove_outliers
```

```python
q1 = horses_new.total_protein.quantile(0.25)
q3 = horses_new.total_protein.quantile(0.75)
print(q1, q3)
iqr = q3 - q1
lower_bound = q1 - 0.1*iqr
upper_bound = q3 + (2*iqr)
print(lower_bound, upper_bound)
remove_outliers = horses_new[~horses_new.total_protein.between(lower_bound, upper_bound, inclusive=True)].sort_values('total_protein', ascending = True)
remove_outliers
```

Температура - есть несколько выбросов по границам темепратурного интервала, но они не критические, я считаю, что вполне может быть, что у лошади была температура 35 или 40 градусов, при нормально температуре тела 37,8, поэтому мы их оставляем

Пульс - присутвует несколько выбросов, пульс за пределом верхней границы, но не намного, причем у всех лошадей с таким пульсом была высокая температура, вероятно она была в шоковом состоянии, оставляем.

Уровень белка -если брать формулу для границ, как на занятии, то выбросов нет. Но мы видим, что имеем огромную разницу между первым и третим квантилем, и также большую меру разброса. Возможно при сильном обезвоживании он может достигиать высоких значений, оставляем.

Пропущенные значения в Surgery и outcome можно заменить модой, чаще всего встречающееся значение и это всего один случай в нашем дата сете.

rectal_temperature - ранее мы видели, что все средние у нас более менее одинаковы, поэто пропущенное значение заменим на среднее - 38.2.

abdominal_distension - можно заметить, что если лошадь умерла или была усыплена (outcome = 3 или 2), то показатель вздутия живота у животного был 4 или 3, поэтому предлагаю заменить пропущенные значения на медиану с группировкой по outcome.

capillary_refill_time - так же как и у abdominal_distension при сметельных случаях насыщение кислородом было медленным, те показатель равен 2. Заменим им все пропущенные занчения на седиану с группировкой по outcome.

total_protein - не понятно на что заменять, значение скачет и не понятно от чего зависит. Опять же проедлагаю заменить на среднее.

И еще предлагаю удалить все строки, где больше 4х пропущенных значений. (Есть данныу в которых известно только была ли операция и что случилось, остальные данные пропущены, такие данные будем считать статистически не значимеми и можем их удалить)

```python
horses_clean.isna().sum(axis = 1).sort_values()
horses_clean.isna().sum()
```

```python
horses_clean.Surgery.fillna(horses_new.Surgery.mode()[0], inplace=True)
```

```python
horses_clean.outcome.fillna(horses_new.outcome.mode()[0], inplace=True)
```

```python
horses_clean.Age.fillna(horses_new.Age.mode()[0], inplace=True)
```

```python
horses_clean.Puls.fillna(horses_new.Puls.mean(), inplace=True)
```

```python
horses_clean.rectal_temperature.fillna(horses_new.rectal_temperature.mean(), inplace=True)
```

```python
horses_clean.total_protein.fillna(horses_new.total_protein.mean(), inplace=True)
```

```python
horses_clean.capillary_refill_time.fillna(horses_new.groupby(['outcome']).capillary_refill_time.transform('median'), inplace=True)
```
