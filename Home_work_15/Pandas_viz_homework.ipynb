```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from pylab import rcParams
rcParams['figure.figsize'] = 10,10
```

```python
happy_2015 = pd.read_csv('2015.csv')
happy_2015.head()
```

```python
happy_2016 = pd.read_csv('2016.csv', header = 0,
                        names = ['Country', 'Region', 'Rank','Score', 'Low_confidence','High_confidence', 'GDP_per_capita','Family','Health', 'Freedom','Trust', 'Generosity','DR'])
happy_2016.head()
list(happy_2016.columns)
#happy_2019 = happy_2019.rename(columns={"Сountry":"Country","Happiness Rank":"Rank","Happiness Score":"Score","GDP per capita":"Economy","Healthy life expectancy":"Health","Trust (Government Corruption)":"Trust","Dystopia Residual":"Dystopia_Residual"})
#happy_2016.head()
```

```python
happy_2017 = pd.read_csv('2017.csv')
happy_2017.head()
```

```python
happy_2018 = pd.read_csv('2018.csv')
happy_2018 = happy_2018.rename(columns={'Overall rank': "Rank", 'Country or region': "Country"})
happy_2018.head()
```

```python
happy_2018_region = happy_2018.merge(happy_2016[['Country', 'Region']], on='Country').groupby('Region').mean().sort_values('Score')
happy_2018_region = happy_2018_region.reset_index()
happy_2018_region['Region'].astype(str)
```

```python
happy_2019 = pd.read_csv('2019.csv')
happy_2019 = happy_2019.rename(columns={'Overall rank': "Rank", 'Country or region': "Country"})
#list(happy_2019.columns)
#happy_2019.head()country
```

Сделаем разбивку по регионам и посмотрим распределение 'уровня счастья" в 2019 году

```python
happy_2019_region = happy_2019.merge(happy_2016[['Country', 'Region']], on='Country').groupby('Region').mean().sort_values('Score')
happy_2019_region = happy_2019_region.reset_index()
happy_2019_region['Region'].astype(str)
happy_2019_region.info()
```

```python
plt.barh(happy_2019_region['Region'], happy_2019_region['Score'])
plt.xlabel('')
plt.ylabel('')
plt.title('Avergage Happines Score per Region, 2019 ')
#plt.text(20000, 0.5, 'Happiness score ')
#plt.grid()
```

Сравним с 2018 годом

```python
width = 0.25
lables = happy_2019_region['Region']
x = np.arange(len(lables))

fig, ax = plt.subplots()

plt.barh(x-width/2, happy_2019_region['Score'], width, label = '2019')
plt.xlabel('')
plt.ylabel('')
plt.title('Avergage Happines Score per Region, 2018 vs 2019 ')
#plt.grid()

plt.barh(x + width/2, happy_2018_region['Score'], width, label = '2018')
plt.xlabel('')
plt.ylabel('')



#ax.set_ylabel('Scores')
#ax.set_title('Scores by group and gender')
ax.set_yticks(x)
ax.set_yticklabels(lables)
ax.legend()
```
