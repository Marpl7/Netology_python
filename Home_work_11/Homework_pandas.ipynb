Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')

ratings.head()

```

```python
ratings[ratings.rating ==  5].groupby('movieId').count().sort_values('rating', ascending = False).merge(movies, on = 'movieId').filter(items = ['movieId', 'title']).head(1)
```

Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
power.head()
```

```python
pribalt = power[(power['country'] == 'Litva') | (power['country'] == 'Estonia') | (power['country'] == 'Lithuania')]
pribalt_cat_selection = pribalt[ (pribalt['category'] == 4) | (pribalt['category'] == 12) | (pribalt['category'] == 21)]
pribalt_year_selection = pribalt_cat_selection.query('year > 2004').query('year < 2011').query('quantity > 0')
pribalt_year_selection['quantity'].sum()
```
