## как критики относятся к спортивным играм?

H0: средняя оценка критиков для спортивных игр >=7, критикам нравятся спортивные игры
H1: средняя оценка критиков для спортивных игр <7, критикам не нравятся спортивные игры

alfa = 0.05
сравниваем значения выборки с константой - нужен одновыборочный тест
инересует сравнение в меньшую сторону - односторонний тест

````python
from scipy import stats as st
alpha = 0.05
result = st.ttest_1samp(games[games.Genre == 'Sports']['Critic_Score'].dropna(), 7)
print(result)

if (result.statistic < 0) & (result.pvalue / 2 < alpha):
    print('Отвергаем нулевую гипотезу, среднее меньше 7')
else:
    print('Не отвергаем нулевую нулевую гипотезу, критикам нравятся спортивные игры')
````

S > 0, p < 0.05, те с высокой долей вероятности мы можем утверждыть, что критики ставят оценку 7 и больше, те им нравятся спортивные игры

## Критикам нравятся больше игры на PC или на PS4?  

H0: средняя оценка критиков для игрв на PS4 >= средней оценки для игр на PC, PS4 нравится больше
H1: средняя оценка критиков для игрв на PS4 < средней оценки для игр на PC

alfa = 0.05
сравниваем среднее двух выборок
двухвыборочный односторонний тест

````python
result = st.ttest_ind(games[games.Platform == 'PS4']['Critic_Score'].dropna(), games[games.Platform == 'PC']['Critic_Score'].dropna(), equal_var=False)
print(result)
alpha = 0.05
if (result.statistic <0) & (result.pvalue/2 < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу, PS4 нравится больше')
````

S > 0, P < 0.05, c высокой долей вероятности можем утверждать, что вредняя оценка за PS4 выше PC, те PS4 нравится больше

## Критикам нравятся больше стрелялки или стратегии?  

H0: средняя оценка критиков для стрелялок >= средней оценки для стрвтегий, стрелялки нравится больше
H1: средняя оценка критиков для стрелялок < средней оценки для стратегий

alfa = 0.05
сравниваем среднее двух выборок
двухвыборочный односторонний тест

````python
result = st.ttest_ind(games[games.Genre == 'Shooter']['Critic_Score'].dropna(), games[games.Genre == 'Strategy']['Critic_Score'].dropna(), equal_var=False)
print(result)
alpha = 0.05
print(result.pvalue/2)
if (result.statistic <0) & (result.pvalue/2 < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу, стрелялки нравится больше')
````

S < 0, P > 0.05, тут S < 0, те среднее по стрелялкам ниже среднего по стратегиям, но P Value чуть выше 0.05, поэтому мы не можем с высокой долей уверенности говорить, что это статистически значимо, поэтому не отвергаем нулевую гипотезу
Но если повысим альфа до 0.055, то нулевая гипотеза будет отвергнута

````python
import re
import pandas as pd
````

````python
spam = pd.read_csv('Spam.csv')
spam
````

````python
spam_lower_case = [spam.lower() for spam in spam.Message]
spam_lower_case
````

````python
# уберем все служебные символы
only_words_text = [re.sub('[\W_]+',' ', text) for text in spam_lower_case]

only_words_text
````

````python
# разделим все на слова
texts_splitted = [text.split() for text in only_words_text]

texts_splitted
````

````python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
````

````python
from nltk.corpus import stopwords
stopwords_set = set(stopwords.words('english'))
without_sw = [[word for word in text if word not in stopwords_set] for text in texts_splitted]
without_sw
````

````python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
lemmatized_text = [[wordnet_lemmatizer.lemmatize(word) for word in text] for text in without_sw]
lemmatized_text[15]
````

````python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(spam.Message)
names = tfidf.get_feature_names()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=names)
tfidf_matrix
````

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(tfidf_matrix, spam.Category, test_size=0.30, random_state=42)
````

````python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=42)
````

````python
lr.fit(X_train, y_train)
````

````python
y_pred = lr.predict(X_test)
````

````python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
````

````python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
````

True positive - 14446 раз модель правильно предсказала, что это не спам, True Negative - 165 раз модель правильно предсказала, что это спам, False Negative - 59 раз ошиблась, был спам, а она сказала, что не спасм, и False Positive - 2 раза ошиблась, был не спам, а можель сказа, что спам
